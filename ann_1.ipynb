```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define bipolar sigmoid activation function
def bipolar_sigmoid(x):
    return (1 - np.exp(-x)) / (1 + np.exp(-x))
```

```python
# Define x range
x = np.linspace(-10, 10, 100)

# Calculate activation values
sigmoid_y = sigmoid(x)
bipolar_sigmoid_y = bipolar_sigmoid(x)
```

```python
# Plot activation functions
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(x, sigmoid_y, label='Sigmoid', color='b')
plt.title('Sigmoid')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.grid(True)
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(x, relu_y, label='ReLU', color='r')
plt.title('ReLU')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.legend()
```
